```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Œ¶ŒøœÅœÑœéŒΩŒøœÖŒºŒµ œÑŒ± Œ¥ŒµŒ¥ŒøŒºŒ≠ŒΩŒ± Œ±œÄœå œÑŒø Œ±œÅœáŒµŒØŒø
df = pd.read_csv('data.csv')

# ŒüœÅŒØŒ∂ŒøœÖŒºŒµ œÑŒπœÇ œÉœÑŒÆŒªŒµœÇ A, B Œ∫Œ±Œπ C
A = df['A'] # Œ†ŒØŒΩŒ±Œ∫Œ±œÇ œÄŒøœÖ Œ≠œáŒµŒπ œÑŒ± x
B = df['B'] # Œ†ŒØŒΩŒ±Œ∫Œ±œÇ œÄŒøœÖ Œ≠œáŒµŒπ œÑŒ± fŒí
C = df['C'] # Œ†ŒØŒΩŒ±Œ∫Œ±œÇ œÄŒøœÖ Œ≠œáŒµŒπ œÑŒ± fC
```

---
<span style="font-size: larger;">**ŒìœÅŒ±œÜŒπŒ∫ŒÆ Œ†Œ±œÅŒ¨œÉœÑŒ±œÉŒ∑ œÑŒøœÖ $ùëì$**</span><sub>***B***</sub> $(x)$ , <span style="font-size: smaller;">œéœÄŒøœÖ œÑŒø $x$ œÄŒ±ŒØœÅŒΩŒµŒπ œÑŒπŒºŒ≠œÇ Œ±œÄœå œÑŒ∑ŒΩ œÉœÑŒÆŒªŒ∑ Œë</span></br>

```python
plt.scatter(A, B, label='B')

# ŒúŒπŒ± œÄœÅŒøœÉŒ≠Œ≥Œ∫ŒπœÉŒ∑ œÑŒ∑œÇ œÉœÖŒΩŒ¨œÅœÑŒ∑œÉŒ∑œÇ œÄŒøœÖ œàŒ¨œáŒΩŒøœÖŒºŒµ
# (Œ¥ŒµŒΩ ŒÆœÑŒ±ŒΩ Œ±œÄŒ±œÅŒ±ŒØœÑŒ∑œÑŒø ŒΩŒ± Œ≥ŒØŒΩŒµŒπ, ŒºœÄŒøœÅŒøœçŒΩ ŒøŒπ ŒµœÄœåŒºŒµŒΩŒµœÇ 2 Œ≥œÅŒ±ŒºŒºŒ≠œÇ ŒΩŒ± Œ±œÜŒ±ŒπœÅŒµŒ∏ŒøœçŒΩ Œ±ŒΩ Œ¥ŒµŒΩ œáœÅŒµŒπŒ¨Œ∂ŒøŒΩœÑŒ±Œπ)
x = np.linspace(min(A), max(A), 100)
plt.plot(x, 2.25*np.log10(x), color='red', label='log10')

plt.xlabel('A')
plt.ylabel('B')
plt.legend()
plt.show()
```

<span style="font-size: larger;">**Œ£Œ•ŒúŒ†ŒïŒ°ŒëŒ£ŒúŒë**</span></br>
Œ†Œ±œÅŒ±œÑŒ∑œÅŒøœçŒºŒµ œåœÑŒπ Œ≤Œ¨Œ∂ŒøŒΩœÑŒ±œÇ œÑŒπœÇ œÑŒπŒºŒ≠œÇ œÑŒøœÖ Œë œÉœÑŒøŒΩ Œ¨ŒæŒøŒΩŒ± x'x Œ∫Œ±Œπ œÑŒπœÇ œÑŒπŒºŒ≠œÇ œÑŒøœÖ Œí œÉœÑŒøŒΩ Œ¨ŒæŒøŒΩŒ± y'y Œ∫Œ±œÑŒ±ŒªŒÆŒ≥ŒøœÖŒºŒµ œÉŒµ ŒºŒØŒ± œÉœÖŒΩŒ¨œÅœÑŒ∑œÉŒ∑ <span style="font-family: 'Times New Roman'; font-size: 18px;">**ŒªŒøŒ≥Œ±œÅŒπŒ∏ŒºŒπŒ∫Œøœç**</span> œÑœçœÄŒøœÖ.</br>
<span style="font-size: smaller;">Œ†Œ±œÅŒ±Œ∫Œ¨œÑœâ œÜŒ±ŒØŒΩŒµœÑŒ±Œπ Œ≠ŒΩŒ± œÄŒ±œÅŒ¨Œ¥ŒµŒπŒ≥ŒºŒ± Œ≥œÅŒ±œÜŒπŒ∫ŒÆœÇ œÄŒ±œÅŒ¨œÉœÑŒ±œÉŒ∑œÇ ŒºŒπŒ±œÇ <span style="font-family: 'Times New Roman'; font-size: 14px;">**ŒªŒøŒ≥Œ±œÅŒπŒ∏ŒºŒπŒ∫ŒÆœÇ**</span> œÉœÖŒΩŒ¨œÅœÑŒ∑œÉŒ∑œÇ:</span>

```python
x_values = np.linspace(1, 1000, 100)
y_values = np.log10(x_values)

plt.plot(x_values, y_values, color='red', label='log(x)')

plt.xlabel('x')
plt.ylabel('log10(x)')
plt.legend()
plt.show()
```

---
<span style="font-size: larger;">**ŒìœÅŒ±œÜŒπŒ∫ŒÆ Œ†Œ±œÅŒ¨œÉœÑŒ±œÉŒ∑ œÑŒøœÖ $ùëì$**</span><sub>***C***</sub> $(x)$ , <span style="font-size: smaller;">œéœÄŒøœÖ œÑŒø $x$ œÄŒ±ŒØœÅŒΩŒµŒπ œÑŒπŒºŒ≠œÇ Œ±œÄœå œÑŒ∑ŒΩ œÉœÑŒÆŒªŒ∑ Œë</span></br>

```python
plt.scatter(A, C, label='C')

# ŒúŒπŒ± œÄœÅŒøœÉŒ≠Œ≥Œ∫ŒπœÉŒ∑ œÑŒ∑œÇ œÉœÖŒΩŒ¨œÅœÑŒ∑œÉŒ∑œÇ œÄŒøœÖ œàŒ¨œáŒΩŒøœÖŒºŒµ
# (Œ¥ŒµŒΩ ŒÆœÑŒ±ŒΩ Œ±œÄŒ±œÅŒ±ŒØœÑŒ∑œÑŒø ŒΩŒ± Œ≥ŒØŒΩŒµŒπ, ŒºœÄŒøœÅŒøœçŒΩ ŒøŒπ ŒµœÄœåŒºŒµŒΩŒµœÇ 2 Œ≥œÅŒ±ŒºŒºŒ≠œÇ ŒΩŒ± Œ±œÜŒ±ŒπœÅŒµŒ∏ŒøœçŒΩ Œ±ŒΩ Œ¥ŒµŒΩ œáœÅŒµŒπŒ¨Œ∂ŒøŒΩœÑŒ±Œπ)
x = np.linspace(min(A), max(A), 100)
plt.plot(x, np.cbrt(x), color='red', label='cbrt')

plt.xlabel('A')
plt.ylabel('C')
plt.legend()
plt.show()
```

<span style="font-size: larger;">**Œ£Œ•ŒúŒ†ŒïŒ°ŒëŒ£ŒúŒë**</span></br>
Œ†Œ±œÅŒ±œÑŒ∑œÅŒøœçŒºŒµ œåœÑŒπ Œ≤Œ¨Œ∂ŒøŒΩœÑŒ±œÇ œÑŒπœÇ œÑŒπŒºŒ≠œÇ œÑŒøœÖ Œë œÉœÑŒøŒΩ Œ¨ŒæŒøŒΩŒ± x'x Œ∫Œ±Œπ œÑŒπœÇ œÑŒπŒºŒ≠œÇ œÑŒøœÖ C œÉœÑŒøŒΩ Œ¨ŒæŒøŒΩŒ± y'y Œ∫Œ±œÑŒ±ŒªŒÆŒ≥ŒøœÖŒºŒµ œÉŒµ ŒºŒØŒ± œÉœÖŒΩŒ¨œÅœÑŒ∑œÉŒ∑ œÑœçœÄŒøœÖ <span style="font-family: 'Times New Roman'; font-size: 18px;">**Œ∫œÖŒ≤ŒπŒ∫ŒÆœÇ œÅŒØŒ∂Œ±œÇ**</span>.</br>
<span style="font-size: smaller;">Œ†Œ±œÅŒ±Œ∫Œ¨œÑœâ œÜŒ±ŒØŒΩŒµœÑŒ±Œπ Œ≠ŒΩŒ± œÄŒ±œÅŒ¨Œ¥ŒµŒπŒ≥ŒºŒ± Œ≥œÅŒ±œÜŒπŒ∫ŒÆœÇ œÄŒ±œÅŒ¨œÉœÑŒ±œÉŒ∑œÇ œÑŒ∑œÇ <span style="font-family: 'Times New Roman'; font-size: 14px;">**Œ∫œÖŒ≤ŒπŒ∫ŒÆœÇ œÅŒØŒ∂Œ±œÇ**</span>:</span>

```python
x_values = np.linspace(2, 1000, 100)

y_values = np.cbrt(x_values)

plt.plot(x_values, y_values, color='red', label='cbrt(x)')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Œ£œáŒµŒ¥ŒØŒ±œÉŒ∑ ŒöœÖŒ≤ŒπŒ∫ŒÆœÇ Œ°ŒØŒ∂Œ±œÇ œÑŒøœÖ x')
plt.legend()
plt.show()
```
